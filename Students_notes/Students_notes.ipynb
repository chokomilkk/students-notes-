```python
import pandas as pan
```

```python
df = pan.read_excel('students_university_grades.xlsx') #функция панд для загрузки файла эксель в дата фрайм
def grade_text(grade):
    if grade == 1:
        return ("sehr gut")
    elif  grade == 2:
        return ("gut")
    elif grade == 3:
        return ("befriedigend")
    elif grade == 4:
        return ("ausreichend")
    else:
        return("magelhaft")
def student_description(row):
    return f"{row['Name']}, ({row['Faculty']}) - {grade_text(row['Grade'])}"
df['description'] = df.apply(student_description, axis=1)
print(df)
#все это для проверки условий, что бы выводило оценки 
```

```python
avg = df['Grade'].mean()
worst_grade = df['Grade'].max() 
best_grade = df['Grade'].min()
above_avg = df[df['Grade'] < avg]['Name']
print(f"{avg:.2f}")
print(best_grade)
print(worst_grade)
print("-----------------------------------")
print(above_avg)
```
